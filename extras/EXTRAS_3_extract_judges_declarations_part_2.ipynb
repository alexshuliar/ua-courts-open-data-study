```python
# function to compare dates

def compare_changes_date(judge):
    
    print('Checking for judge:', judge)
    
    for idx, dec in enumerate(loaded_aggregations[judge]):

        try:
            changes_dec = dec['declaration']['intro']['doc_type'] == 'Форма змін'
            
        except KeyError:
            changes_dec = False

        if changes_dec:

            changes_year = dec['declaration']['intro']['declaration_year']
            
            if changes_year == 2019:
                
                print('Changes year is 2019')
                
            else:
            
            
                changes_date = pd.to_datetime(dec['declaration']['intro']['date'])
                
                try:

                    annual_dec = [dec for dec in loaded_aggregations[judge] if dec['declaration']['intro']['declaration_year'] == changes_year
                                                                     and dec['declaration']['intro']['doc_type'] == 'Щорічна'][0]
                except IndexError:
                    try:
                        annual_dec = [dec for dec in loaded_aggregations[judge] if dec['declaration']['intro']['declaration_year'] == changes_year
                                                                     and dec['declaration']['intro']['doc_type'] == 'Після звільнення'][0]
                    
                    except IndexError:
                        try:
                            annual_dec = [dec for dec in loaded_aggregations[judge] if dec['declaration']['intro']['declaration_year'] == changes_year
                                                                         and dec['declaration']['intro']['doc_type'] == 'Кандидата на посаду'][0]
                        except IndexError:
                            print(f'No annual declaration for year: {changes_year}')
                            continue

                annual_date = pd.to_datetime(annual_dec['declaration']['intro']['date'])


                condition = annual_date > changes_date


                print(f'------ changes: {changes_date}, annual_pub: {annual_date} -- Changes is earlier: {condition}')
```

```python
# there are several ways to come up with dataframe from our json data
# I've ended up with the function because only this one works in those entries where there is a mess in ordering of keys

def generate_df(aggregations_list):
    
    new_dict = {}


    for dec in aggregations_list:
        for key, value in dec.items():
            new_dict.setdefault(key, []).append(value)

    rotated_df = pd.DataFrame(new_dict, index=list(map(int, new_dict['year'])))

    return rotated_df.transpose()
```

```python
def summarize_df(dataframe, kind='max'):
    
    if dataframe.name == 'vehicles.max_year':
        dataframe = dataframe.replace('!немає авто', 0).astype(int)

        return dataframe.max()

    if dataframe.name == 'vehicles.all_names':
        dataframe = dataframe.apply(lambda x: x.split(';') if type(x) == str else x).apply(lambda x: 0 if x == [''] else len(x))

        return dataframe.max()


    if dataframe.name in ['assets_incomes_ratio', 'liabilities_assets_and_incomes_ratio',
                                    'expenses_assets_and_incomes_ratio', 'assets.family_ratio_cat',
                                    'incomes.family_ratio_cat', 'estate.family_land_ratio_cat', 
                                    'estate.family_other_ratio_cat']:

        dataframe = dataframe.str.get(0)

        return dataframe.max()



    if type(dataframe.iloc[0]) == bool:
        return dataframe.any()

    elif type(dataframe.iloc[0]) == float or type(dataframe.iloc[0]) == int:

        return dataframe.max()
```

```python
def get_df_from_json(json_file):
    
    judges_dataframes = []
    
    for judge in json_file.keys():

        declarations_list = json_file[judge]


        aggregated_data = []


        for dec in declarations_list:

            if 'aggregated' in dec['declaration'].keys():

                aggregated_data.append(dec['declaration']['aggregated'])
                
        
        if aggregated_data == []:
            
#             print(judge)

            continue

        accum_df = generate_df(aggregated_data)
        
        basic_features = accum_df.loc[['cash_flag_500k', 'assets.cash.total', 'assets.declarant' , 'assets.family',
                                    'assets.total', 'expenses.total', 'liabilities.total', 'estate.declarant_land', 
                                    'assets.family_ratio', 'estate.declarant_other', 'estate.family_land', 'estate.total_land',
                                    'estate.family_other', 'estate.total_other', 'estate.family_land_ratio',
                                   'estate.family_other_ratio', 'vehicles.total_cost', 'vehicles.max_year', 
                                    'vehicles.all_names', 'incomes.declarant', 'incomes.family', 'incomes.total',
                                   'incomes.family', 'incomes.family_ratio',
                                   'assets_incomes_ratio', 'liabilities_assets_and_incomes_ratio',
                                    'expenses_assets_and_incomes_ratio', 'assets.family_ratio_cat',
                                    'incomes.family_ratio_cat', 'estate.family_land_ratio_cat', 
                                    'estate.family_other_ratio_cat']]
        
        reduced_series = basic_features.apply(summarize_df, axis=1)
            
        
        # NOTE: here we changed zero with the small eps value, 
        # not sure that it's a best way to handle this, but started with it
        pct_change = basic_features.loc[['assets.total', 
                    'incomes.total']].replace(0, np.finfo(np.float).eps).pct_change(axis=1).mean(axis=1).fillna(0).rename({'assets.total': 'assets_pct_change', 
                                                                              'incomes.total': 'incomes_pct_change'})
        
        final_series = pd.concat([reduced_series, pct_change])
        
        judge_row = pd.DataFrame(final_series, columns=[judge]).transpose()

        
        judges_dataframes.append(judge_row)
    
    concatenated_df = pd.concat(judges_dataframes)
    
    
    return concatenated_df
```
